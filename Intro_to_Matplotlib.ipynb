```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
student_details={
    "Names":["Kyle","Stacey","Bob","Ken","Hana","William","Theo","Michael","Liam","Kris"],
    "Scores":[65,44,80,81,54,100,98,79,99,79]
}
```

```python
print(student_details)
```

```python
sd=pd.DataFrame(student_details)
sd
```

```python
plt.bar(sd['Names'],sd['Scores'])
plt.title("Student Scores")
plt.xlabel("Names")
plt.ylabel("Scores")
plt.xticks(rotation=45)
plt.show()
```

#Findings

* **William** is  the **top** performing student because his bar is the **highest**

* **Kris'** and **Michael's** bars have the **same height**

* **Stacey** is the **lowest** performing student because her bar is the **lowest**

```python
plt.plot(sd['Names'],sd['Scores'])
plt.title("Student Scores Trend")
plt.xlabel("Names")
plt.ylabel("Scores")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# Findings
* **The class'** performance was *irregular* because the graph wasn't **in a specific patern**

```python
sd_sorted=sd.sort_values(by="Scores")
sd_sorted
```

```python
plt.plot(sd_sorted['Names'],sd_sorted['Scores'])
plt.title("Student Scores Trend")
plt.xlabel("Names")
plt.ylabel("Scores")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
